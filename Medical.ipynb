```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/insurance.csv')
```

```python
df.head()
df.info()
```

Podemos observar que o dataframe possui 7 variáveis:



*   age: idade do paciente
*   sex: sexo do paciente
*   bmi: índice de massa corporal (IMC) do paciente
*   children: número de filhos dependentes do paciente
*   smoker: indica se o paciente é fumante ou não
*   region: região de residência do paciente
*   charges: custos médicos individuais cobrados pelo seguro saúde
*   Item da lista

Podemos também verificar a quantidade de dados faltantes no dataframe:

```python
df.isnull().sum()
```

```python
df.describe()
```

Podemos observar que as idades variam entre 18 e 64 anos, com uma média de 39 anos. 

O índice de massa corporal (IMC) varia de 15.96 a 53.13, com uma média de 30.66. A maioria dos pacientes não tem filhos dependentes (cerca de 57%), mas alguns pacientes têm até 5 filhos dependentes. Os custos médicos variam de cerca de $1,121 a $63,770, com uma média de $13,270.

Podemos visualizar a distribuição dos custos médicos por meio de um histograma:

```python
sns.histplot(df['charges'])
plt.title('Distribuição dos custos médicos')
plt.xlabel('Custos')
plt.show()
```

Podemos observar que a distribuição dos custos médicos é assimétrica positiva, indicando que a maioria dos custos médicos é baixa, mas alguns pacientes apresentam custos muito altos.

Podemos também verificar a relação entre as variáveis por meio de gráficos e tabelas. Por exemplo, podemos verificar a relação entre o sexo do paciente e os custos médicos por meio de um boxplot:

```python
sns.boxplot(x='sex', y='charges', data=df)
plt.title('Custos médicos por sexo')
plt.xlabel('Sexo')
plt.ylabel('Custos')
plt.show()
```

Podemos observar que os custos médicos são um pouco mais altos para os homens do que para as mulheres, mas a diferença não é muito grande.

Podemos verificar a relação entre o índice de massa corporal (IMC) e os custos médicos por meio de um scatterplot:

```python
sns.scatterplot(x='bmi', y='charges', data=df)
plt.title('Custos médicos por IMC')
plt.xlabel('IMC')
plt.ylabel('Custos')
plt.show()
```

Podemos observar que há uma tendência positiva entre o IMC e os custos médicos, indicando que pacientes com IMC mais elevado tendem a apresentar custos médicos mais elevados.

Essa foi apenas uma breve análise descritiva dos dados da base "Medical

```python
sns.histplot(df['charges'], kde=True)
plt.title('Distribuição dos custos médicos')
plt.xlabel('Custos médicos')
plt.show()
```

A figura resultante mostra a distribuição dos custos médicos na base de dados. Podemos observar que a distribuição dos custos médicos é assimétrica positiva, o que significa que a maioria dos custos médicos é baixa, mas há alguns pacientes com custos muito altos. A maioria dos custos médicos está concentrada abaixo de $20,000.

A idade do paciente pode ter uma influência significativa no valor dos custos médicos, uma vez que o risco de desenvolver certas condições médicas aumenta com a idade.

Podemos visualizar a relação entre a idade e os custos médicos por meio de um scatterplot:

```python
sns.scatterplot(x='age', y='charges', data=df)
plt.title('Custos médicos por idade')
plt.xlabel('Idade')
plt.ylabel('Custos')
plt.show()
```

Podemos observar que há uma tendência positiva entre a idade e os custos médicos, indicando que pacientes mais velhos tendem a apresentar custos médicos mais elevados.

Podemos também dividir a base de dados em grupos etários e verificar a distribuição dos custos médicos em cada grupo:

```python
age_bins = [18, 30, 40, 50, 60, 70, 80]
age_labels = ['18-29', '30-39', '40-49', '50-59', '60-69', '70+']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)

sns.boxplot(x='age_group', y='charges', data=df)
plt.title('Custos médicos por grupo etário')
plt.xlabel('Grupo etário')
plt.ylabel('Custos')
plt.show()
```

Podemos observar que os custos médicos aumentam em média para cada grupo etário mais velho, com o grupo etário de 50-59 anos apresentando a maior mediana de custos médicos. Isso sugere que a idade é uma variável importante a ser considerada na análise dos custos médicos.

Para comparar a influência das variáveis idade, sexo, IMC, fumante ou não e região de residência no preço dos custos médicos, podemos criar gráficos de boxplot para cada variável.

```python
fig, axs = plt.subplots(ncols=2, nrows=3, figsize=(12,12))

sns.boxplot(x='sex', y='charges', data=df, ax=axs[0][0])
sns.boxplot(x='smoker', y='charges', data=df, ax=axs[0][1])
sns.boxplot(x='region', y='charges', data=df, ax=axs[1][0])
sns.scatterplot(x='bmi', y='charges', data=df, ax=axs[1][1])
sns.scatterplot(x='age', y='charges', data=df, ax=axs[2][0])

age_bins = [18, 30, 40, 50, 60, 70, 80]
age_labels = ['18-29', '30-39', '40-49', '50-59', '60-69', '70+']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)
sns.boxplot(x='age_group', y='charges', data=df, ax=axs[2][1])

plt.tight_layout()
plt.show()
```

O gráfico gerado apresenta 6 subplots, cada um comparando uma variável com o preço dos custos médicos:

O primeiro subplot compara o sexo com o preço dos custos médicos, mostrando que a mediana dos custos médicos é maior para homens do que para mulheres.


O segundo subplot compara se o paciente é fumante ou não com o preço dos custos médicos, mostrando que fumantes apresentam custos médicos muito maiores do que não fumantes.


O terceiro subplot compara a região de residência com o preço dos custos médicos, mostrando que a região sul apresenta a maior mediana de custos médicos.


O quarto subplot é um scatterplot que compara o IMC com o preço dos custos médicos, mostrando que há uma tendência positiva entre o IMC e os custos médicos, indicando que pacientes com IMC mais elevado apresentam custos médicos mais elevados.


O quinto subplot é um scatterplot que compara a idade com o preço dos custos médicos, mostrando uma tendência positiva entre a idade e os custos médicos, indicando que pacientes mais velhos tendem a apresentar custos médicos mais elevados.


O sexto subplot compara grupos etários com o preço dos custos médicos, mostrando que a mediana dos custos médicos aumenta em média para cada grupo etário mais velho.


Podemos observar que o status de fumante e o IMC parecem ter uma influência significativa nos custos médicos, enquanto que a idade, sexo e região de residência apresentam uma influência menor.
