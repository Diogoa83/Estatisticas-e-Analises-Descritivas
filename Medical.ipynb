```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/insurance.csv')
```

```python
df.head()
df.info()
```

Podemos observar que o dataframe possui 7 variveis:



*   age: idade do paciente
*   sex: sexo do paciente
*   bmi: ndice de massa corporal (IMC) do paciente
*   children: nmero de filhos dependentes do paciente
*   smoker: indica se o paciente  fumante ou no
*   region: regio de residncia do paciente
*   charges: custos mdicos individuais cobrados pelo seguro sade
*   Item da lista

Podemos tambm verificar a quantidade de dados faltantes no dataframe:

```python
df.isnull().sum()
```

```python
df.describe()
```

Podemos observar que as idades variam entre 18 e 64 anos, com uma mdia de 39 anos. 

O ndice de massa corporal (IMC) varia de 15.96 a 53.13, com uma mdia de 30.66. A maioria dos pacientes no tem filhos dependentes (cerca de 57%), mas alguns pacientes tm at 5 filhos dependentes. Os custos mdicos variam de cerca de $1,121 a $63,770, com uma mdia de $13,270.

Podemos visualizar a distribuio dos custos mdicos por meio de um histograma:

```python
sns.histplot(df['charges'])
plt.title('Distribuio dos custos mdicos')
plt.xlabel('Custos')
plt.show()
```

Podemos observar que a distribuio dos custos mdicos  assimtrica positiva, indicando que a maioria dos custos mdicos  baixa, mas alguns pacientes apresentam custos muito altos.

Podemos tambm verificar a relao entre as variveis por meio de grficos e tabelas. Por exemplo, podemos verificar a relao entre o sexo do paciente e os custos mdicos por meio de um boxplot:

```python
sns.boxplot(x='sex', y='charges', data=df)
plt.title('Custos mdicos por sexo')
plt.xlabel('Sexo')
plt.ylabel('Custos')
plt.show()
```

Podemos observar que os custos mdicos so um pouco mais altos para os homens do que para as mulheres, mas a diferena no  muito grande.

Podemos verificar a relao entre o ndice de massa corporal (IMC) e os custos mdicos por meio de um scatterplot:

```python
sns.scatterplot(x='bmi', y='charges', data=df)
plt.title('Custos mdicos por IMC')
plt.xlabel('IMC')
plt.ylabel('Custos')
plt.show()
```

Podemos observar que h uma tendncia positiva entre o IMC e os custos mdicos, indicando que pacientes com IMC mais elevado tendem a apresentar custos mdicos mais elevados.

Essa foi apenas uma breve anlise descritiva dos dados da base "Medical

```python
sns.histplot(df['charges'], kde=True)
plt.title('Distribuio dos custos mdicos')
plt.xlabel('Custos mdicos')
plt.show()
```

A figura resultante mostra a distribuio dos custos mdicos na base de dados. Podemos observar que a distribuio dos custos mdicos  assimtrica positiva, o que significa que a maioria dos custos mdicos  baixa, mas h alguns pacientes com custos muito altos. A maioria dos custos mdicos est concentrada abaixo de $20,000.

A idade do paciente pode ter uma influncia significativa no valor dos custos mdicos, uma vez que o risco de desenvolver certas condies mdicas aumenta com a idade.

Podemos visualizar a relao entre a idade e os custos mdicos por meio de um scatterplot:

```python
sns.scatterplot(x='age', y='charges', data=df)
plt.title('Custos mdicos por idade')
plt.xlabel('Idade')
plt.ylabel('Custos')
plt.show()
```

Podemos observar que h uma tendncia positiva entre a idade e os custos mdicos, indicando que pacientes mais velhos tendem a apresentar custos mdicos mais elevados.

Podemos tambm dividir a base de dados em grupos etrios e verificar a distribuio dos custos mdicos em cada grupo:

```python
age_bins = [18, 30, 40, 50, 60, 70, 80]
age_labels = ['18-29', '30-39', '40-49', '50-59', '60-69', '70+']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)

sns.boxplot(x='age_group', y='charges', data=df)
plt.title('Custos mdicos por grupo etrio')
plt.xlabel('Grupo etrio')
plt.ylabel('Custos')
plt.show()
```

Podemos observar que os custos mdicos aumentam em mdia para cada grupo etrio mais velho, com o grupo etrio de 50-59 anos apresentando a maior mediana de custos mdicos. Isso sugere que a idade  uma varivel importante a ser considerada na anlise dos custos mdicos.

Para comparar a influncia das variveis idade, sexo, IMC, fumante ou no e regio de residncia no preo dos custos mdicos, podemos criar grficos de boxplot para cada varivel.

```python
fig, axs = plt.subplots(ncols=2, nrows=3, figsize=(12,12))

sns.boxplot(x='sex', y='charges', data=df, ax=axs[0][0])
sns.boxplot(x='smoker', y='charges', data=df, ax=axs[0][1])
sns.boxplot(x='region', y='charges', data=df, ax=axs[1][0])
sns.scatterplot(x='bmi', y='charges', data=df, ax=axs[1][1])
sns.scatterplot(x='age', y='charges', data=df, ax=axs[2][0])

age_bins = [18, 30, 40, 50, 60, 70, 80]
age_labels = ['18-29', '30-39', '40-49', '50-59', '60-69', '70+']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)
sns.boxplot(x='age_group', y='charges', data=df, ax=axs[2][1])

plt.tight_layout()
plt.show()
```

O grfico gerado apresenta 6 subplots, cada um comparando uma varivel com o preo dos custos mdicos:

O primeiro subplot compara o sexo com o preo dos custos mdicos, mostrando que a mediana dos custos mdicos  maior para homens do que para mulheres.


O segundo subplot compara se o paciente  fumante ou no com o preo dos custos mdicos, mostrando que fumantes apresentam custos mdicos muito maiores do que no fumantes.


O terceiro subplot compara a regio de residncia com o preo dos custos mdicos, mostrando que a regio sul apresenta a maior mediana de custos mdicos.


O quarto subplot  um scatterplot que compara o IMC com o preo dos custos mdicos, mostrando que h uma tendncia positiva entre o IMC e os custos mdicos, indicando que pacientes com IMC mais elevado apresentam custos mdicos mais elevados.


O quinto subplot  um scatterplot que compara a idade com o preo dos custos mdicos, mostrando uma tendncia positiva entre a idade e os custos mdicos, indicando que pacientes mais velhos tendem a apresentar custos mdicos mais elevados.


O sexto subplot compara grupos etrios com o preo dos custos mdicos, mostrando que a mediana dos custos mdicos aumenta em mdia para cada grupo etrio mais velho.


Podemos observar que o status de fumante e o IMC parecem ter uma influncia significativa nos custos mdicos, enquanto que a idade, sexo e regio de residncia apresentam uma influncia menor.
