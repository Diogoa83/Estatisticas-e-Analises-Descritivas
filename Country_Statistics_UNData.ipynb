```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o dataset
df = pd.read_csv('/content/country_profile_variables.csv')


```

```python

df['Individuals using the Internet (per 100 inhabitants)'] = df['Individuals using the Internet (per 100 inhabitants)'].astype(float)
```

```python


df['GDP growth rate (annual %, const. 2005 prices)'] = pd.to_numeric(df['GDP growth rate (annual %, const. 2005 prices)'], errors='coerce')
df['Economy: Agriculture (% of GVA)'] = pd.to_numeric(df['Economy: Agriculture (% of GVA)'], errors='coerce')
df['Employment: Agriculture (% of employed)'] = pd.to_numeric(df['Employment: Agriculture (% of employed)'], errors='coerce')
df['Employment: Industry (% of employed)'] = pd.to_numeric(df['Employment: Industry (% of employed)'], errors='coerce')
df['Employment: Services (% of employed)'] = pd.to_numeric(df['Employment: Services (% of employed)'], errors='coerce')
df['Unemployment (% of labour force)'] = pd.to_numeric(df['Unemployment (% of labour force)'], errors='coerce')
df['Labour force participation (female/male pop. %)'] = pd.to_numeric(df['Labour force participation (female/male pop. %)'], errors='coerce')
df['International trade: Exports (million US$)'] = pd.to_numeric(df['International trade: Exports (million US$)'], errors='coerce')
df['International trade: Imports (million US$)'] = pd.to_numeric(df['International trade: Imports (million US$)'], errors='coerce')
df['International trade: Balance (million US$)'] = pd.to_numeric(df['International trade: Balance (million US$)'], errors='coerce')
df['Balance of payments, current account (million US$)'] = pd.to_numeric(df['Balance of payments, current account (million US$)'], errors='coerce')
df['Population growth rate (average annual %)'] = pd.to_numeric(df['Population growth rate (average annual %)'], errors='coerce')
df['Urban population growth rate (average annual %)'] = pd.to_numeric(df['Urban population growth rate (average annual %)'], errors='coerce')
df['Fertility rate, total (live births per woman)'] = pd.to_numeric(df['Fertility rate, total (live births per woman)'], errors='coerce')
df['Life expectancy at birth (females/males, years)'] = pd.to_numeric(df['Life expectancy at birth (females/males, years)'], errors='coerce')
#df['Population age distribution (0-14 / 60+ years, %)'] = pd.to_numeric(df['Population age distribution (0-14 / 60+ years, %)'], errors='coerce')
df['International migrant stock (000/% of total pop.)'] = pd.to_numeric(df['International migrant stock (000/% of total pop.)'], errors='coerce')
df['Refugees and others of concern to UNHCR (in thousands)'] = pd.to_numeric(df['Refugees and others of concern to UNHCR (in thousands)'], errors='coerce')
df['Infant mortality rate (per 1000 live births'] = pd.to_numeric(df['Infant mortality rate (per 1000 live births'], errors='coerce')
```

```python
df['Health: Physicians (per 1000 pop.)'] = pd.to_numeric(df['Health: Physicians (per 1000 pop.)'], errors='coerce')
df['Education: Government expenditure (% of GDP)'] = pd.to_numeric(df['Education: Government expenditure (% of GDP)'], errors='coerce')
df['Education: Primary gross enrol. ratio (f/m per 100 pop.)'] = pd.to_numeric(df['Education: Primary gross enrol. ratio (f/m per 100 pop.)'], errors='coerce')
df['Education: Secondary gross enrol. ratio (f/m per 100 pop.)'] = pd.to_numeric(df['Education: Secondary gross enrol. ratio (f/m per 100 pop.)'], errors='coerce')
df['Education: Tertiary gross enrol. ratio (f/m per 100 pop.)'] = pd.to_numeric(df['Education: Tertiary gross enrol. ratio (f/m per 100 pop.)'], errors='coerce')
df['Mobile-cellular subscriptions (per 100 inhabitants)'] = pd.to_numeric(df['Mobile-cellular subscriptions (per 100 inhabitants)'], errors='coerce')
df['Mobile-cellular subscriptions (per 100 inhabitants).1'] = pd.to_numeric(df['Mobile-cellular subscriptions (per 100 inhabitants).1'], errors='coerce')
df['Forested area (% of land area)'] = pd.to_numeric(df['Forested area (% of land area)'], errors='coerce')
df['CO2 emission estimates (million tons/tons per capita)'] = pd.to_numeric(df['CO2 emission estimates (million tons/tons per capita)'], errors='coerce')
```

```python
# Selecionar os dados para os grficos
df_industrial = df[['country', 'Economy: Industry (% of GVA)']]
#df_industrial = df_industrial.dropna()

df_agriculture = df[['country', 'Economy: Agriculture (% of GVA)']]
#df_agriculture = df_agriculture.dropna()

# Grfico de barras para produo industrial
df_industrial = df_industrial.sort_values(by=['Economy: Industry (% of GVA)'], ascending=False).head(10)
plt.bar(df_industrial['country'], df_industrial['Economy: Industry (% of GVA)'])
plt.xticks(rotation=45)
plt.xlabel('Pas')
plt.ylabel('Produo Industrial (% do PIB)')
plt.title('Produo Industrial dos 10 pases do Mundo com maior PIB industrial')
plt.show()

# Grfico de barras para agronegcio
df_agriculture = df_agriculture.sort_values(by=['Economy: Agriculture (% of GVA)'], ascending=False).head(10)
plt.bar(df_agriculture['country'], df_agriculture['Economy: Agriculture (% of GVA)'])
plt.xticks(rotation=45)
plt.xlabel('Pas')
plt.ylabel('Agricultura (% do PIB)')
plt.title('Agronegcio dos 10 pases do Mundo com maior PIB agrcola')
plt.show()
```

```python
# Selecionar as variveis relevantes
data = df[['country', 'GDP per capita (current US$)', 'Economy: Agriculture (% of GVA)', 'Economy: Industry (% of GVA)']]

# Filtrar os pases com maior produo agrcola
data_agri = data.sort_values(by='Economy: Agriculture (% of GVA)', ascending=False).head(10)

# Filtrar os pases com maior produo industrial
data_ind = data.sort_values(by='Economy: Industry (% of GVA)', ascending=False).head(10)

# Imprimir os resultados
print('Pases com maior produo agrcola:')
print(data_agri)
print('\nPases com maior produo industrial:')
print(data_ind)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados de produo agrcola
countries_agri = data_agri['country']
gdp_agri = data_agri['GDP per capita (current US$)']
agri_percent = data_agri['Economy: Agriculture (% of GVA)']

# Dados de produo industrial
countries_ind = data_ind['country']
gdp_ind = data_ind['GDP per capita (current US$)']
ind_percent = data_ind['Economy: Industry (% of GVA)']

# Grfico de barras para produo agrcola
plt.bar(np.arange(len(countries_agri)), agri_percent, align='center')
plt.xticks(np.arange(len(countries_agri)), countries_agri, rotation=45)
plt.ylabel('Economy: Agriculture (% of GVA)')
plt.title('Pases com maior produo agrcola')

plt.show()

# Grfico de barras para produo industrial
plt.bar(np.arange(len(countries_ind)), ind_percent, align='center')
plt.xticks(np.arange(len(countries_ind)), countries_ind, rotation=45)
plt.ylabel('Economy: Industry (% of GVA)')
plt.title('Pases com maior produo industrial')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Extract economic indicators for Brazil and other countries in the region
economic_indicators = df.loc[df['Region'] == general_info['Region'].values[0], 
                             ['country', 'GDP per capita (current US$)']]
economic_indicators = economic_indicators.sort_values('GDP per capita (current US$)')

# Create a horizontal bar chart to compare GDP per capita of Brazil with other countries in the same region
plt.figure(figsize=(8,6))
plt.barh(economic_indicators['country'], economic_indicators['GDP per capita (current US$)'], color='blue')
plt.barh('Brazil', economic_indicators.loc[economic_indicators['country']=='Brazil', 'GDP per capita (current US$)'].values, 
         color='red')
plt.xlabel('GDP per capita (current US$)')
plt.title('Comparison of GDP per capita (current US$) in Brazil with other countries in the same region')
plt.show()
```

```python
# Selecionar apenas os pases da Amrica Latina
latam = ['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Costa Rica', 'Cuba', 'Ecuador', 'El Salvador', 'Mexico', 'Nicaragua', 
         'Paraguay', 'Peru', 'Uruguay', 'Venezuela']
df_latam = df[df['country'].isin(latam)]

# Selecionar a coluna "Urban population (% of total population)"
urban_population = df_latam[['country', 'Urban population (% of total population)']]

# Calcular a mdia da populao urbana para os pases da Amrica Latina
mean_urban_population = urban_population.groupby('country').mean()

# Criar um grfico de barras para comparar a populao urbana mdia na Amrica Latina
plt.figure(figsize=(16, 6))
plt.bar(mean_urban_population.index, mean_urban_population['Urban population (% of total population)'], color='blue')
plt.title('Average Urban Population in Latin America')
plt.xlabel('Country')
plt.ylabel('Urban Population (% of Total Population)')

# Adicionar uma segunda barra para o Brasil, destacando-o em vermelho
plt.bar('Brazil', mean_urban_population.loc['Brazil', 'Urban population (% of total population)'], color='red')

plt.show()
```
